```python
import numpy as np
from scipy.integrate import solve_ivp
from plotly.subplots import make_subplots
```

```python
q = 100. #L/min
Cain = 1. #mol/L
Tin = 350. #K
V = 100. #L
rho = 1e3 #g/L
Cp = 0.239 #J/(g K)
negHr = 5e4 #J/mol
ER = 8750. #K
k0 = 7.2e10 #1/min
UA = 5e4 #J/(min K)
Tc = 305. #K

Ca0 = 0.5 #mol/L
T0 = 350. #K
tend= 10. #min
```

```python
def rhs(t, vec):
    Ca, T = vec
    k = k0*np.exp(-ER/T)
    dCa = q*(Cain - Ca)/V - k*Ca
    dT = q*(Tin - T)/V - UA/(rho*V*Cp) * (T-Tc) + negHr*k*Ca/(rho*Cp)
    return np.array([dCa, dT])
```

```python
res=solve_ivp(rhs, (0,tend), np.array([Ca0, T0]), method='Radau', dense_output=True)
```

```python
time_plot = np.linspace(0,tend,500)
Caplot, Tplot = res.sol(time_plot)
fig = make_subplots(rows=1, cols=2)
fig.add_scatter(x=time_plot, y=Caplot, name='Ca',row=1,col=1)
fig.add_scatter(x=time_plot, y=Tplot, name='T', row=1, col=2)
fig.update_layout(width=1000, height=500, template='plotly_dark')
```
