```python
import numpy as np
from plotly.subplots import make_subplots
from scipy.integrate import solve_ivp
```

```python
Ah = 0.01 #m3
Ac = 1. #m3
g = 9.81 #m/s2
qin0 = 0.1 #m3/s
h0 = ((qin0/Ah)**2)/(2*g)
tend = 5000. #s
Kc=0.005 #ultimate gain
h3sp = h0*1.1
taui=50
taud=5000
errint0=0
```

```python
def rhs(t, v):
    h1,h2,h3, errint =v
    e=h3sp-h3
    q12 = Ah*(2*g*h1)**0.5
    q23 = Ah*(2*g*h2)**0.5
    q3 = Ah*(2*g*h3)**0.5
    qin = Kc*(e + errint/taui + taud*(q3-q23)/Ac) + qin0

    return np.array([(qin-q12)/Ac, (q12-q23)/Ac, (q23-q3)/Ac, e])
```

```python
res=solve_ivp(rhs, (0,tend), np.array([h0,h0,h0, errint0]), method='Radau', dense_output=True,rtol=1e-9, atol=1e-12)
tplot = np.linspace(0,tend,500)
h1,h2,h3, errint=res.sol(tplot)
fig=make_subplots()
fig.add_scatter(x=tplot,y=h1, name='tank1')
fig.add_scatter(x=tplot,y=h2, name='tank2')
fig.add_scatter(x=tplot,y=h3, name='tank3')
fig.update_layout(width=600,height=400,template='plotly_dark')
```
